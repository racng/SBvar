```python
import sbvar as sb
import tellurium as te
```

```python
ant_model = '''
    J0: S1 -> S2; k1*S1;
    J1: S2 -> S3; k2*S2;

    k1= 0.1; k2 = 0.2;
    S1 = 10; S2 = 0; S3 = 0;
'''
rr = te.loada(ant_model)
exp = sb.experiment.OneWayExperiment(rr, param='S1', range=(0,40), num=50, 
    start=0, end=100, points=10)
```
